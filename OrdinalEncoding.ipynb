```python
import numpy as np
import pandas as pd 
import seaborn as sns
from category_encoders import OrdinalEncoder
```

# Data

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

# Ordinal encoding

- encoding

```python
object_col = titanic.dtypes[titanic.dtypes == "object"].index.tolist() # 범주형 컬럼
print(object_col)
```

```python
# encoder 생성 
ordinal_encoder = OrdinalEncoder(object_col)
ordinal_encoder.fit(titanic[object_col])

# encoding
titanic[object_col] = ordinal_encoder.transform(titanic[object_col])
```

```python
titanic.head()
```

- decoding

```python
titanic[object_col] = ordinal_encoder.inverse_transform(titanic[object_col])
```
