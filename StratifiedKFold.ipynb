# Library

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
import numpy as np
import pandas as pd
```

# Data

```python
iris = load_iris()
features = iris.data
label = iris.target
```

# StratifiedKFold

```python
dt_clf = DecisionTreeClassifier(random_state=156) # 모델 생성
skfold = StratifiedKFold(n_splits=3) # StratifiedKFold 생성
cv_accuracy=[] # fold 별 accuracy 저장
n_iter=0

for train_index, test_index  in skfold.split(features, label):
    # train, test data 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]

    # 학습, 예측
    dt_clf.fit(X_train , y_train)    
    pred = dt_clf.predict(X_test)

    # fold 별 accuracy 
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
    cv_accuracy.append(accuracy)
    print('#{0} 학습 레이블 데이터 분포:\n'.format(n_iter), pd.Series(y_train).value_counts())
    print('#{0} 검증 레이블 데이터 분포:\n'.format(n_iter), pd.Series(y_test).value_counts())
    
# accuracy 평균 list, accuracy 평균 
print('\n## 교차 검증별 정확도:', np.round(cv_accuracy, 4))
print('## 평균 검증 정확도:', np.mean(cv_accuracy)) 
```
